```python
from numpy import random

x = random.normal(size=(2, 3))

print(x)
```

```python
x = random.normal(loc=1, scale=2, size=(2, 3))

print(x)
```

```python
x = random.randn(2, 3)
print(x)
```

## 3\. **Key Differences**

| Function | Mean (loc) | Std Dev (scale) | Parameters |
| --- | --- | --- | --- |
| **`np.random.randn()`** | Fixed at **0** | Fixed at **1** | Only shape |
| **`np.random.normal()`** | Customizable | Customizable | loc, scale, size |

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.normal(size=1000), kind="kde")

plt.show()
```
